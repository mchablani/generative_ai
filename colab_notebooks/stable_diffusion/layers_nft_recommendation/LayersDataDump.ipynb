```python
import csv
import requests
from pathlib import Path
```

```python
output_path = Path("/content/drive/MyDrive/expa/layers_data/images/temp/")
with open("/content/drive/MyDrive/expa/layers_data/layer-nfts-sample-data.csv", newline='') as csvfile:
  reader = csv.DictReader(csvfile)

  row_index = 0
  for row in reader:
    row_index += 1
    # print(row["media"])
    if row["media_type"] == "static":
      url = row["media"]
      # print(url)
      try: 
        # Send a request to download the image
        response = requests.get(url)
        # Extract the filename from the URL
        filename = url.split('/')[-1]
        # print(filename)
        # Save the image to a file
        if len(response.content) > 100:
          with open(output_path / str(row_index), 'wb') as f:
              f.write(response.content)
        else: 
          print(row["media_type"], url, "content too short", len(response.content))  
      except Exception as e:
        print(row["media_type"], url, e)
    else:      
      print(row["media_type"], row["media"])
      continue

    # break

```

```python
import os
from PIL import Image
import shutil
```

```python
folder_path = '/content/drive/MyDrive/expa/layers_data/images/temp/'
dest_folder_path = '/content/drive/MyDrive/expa/layers_data/images/nfts/'

for file_name in os.listdir(folder_path):
# Check if the file is a JPEG image
      file_path = os.path.join(folder_path, file_name)
      try:
          # Open the image file with PIL
          with Image.open(file_path) as img:
              # Do something with the image here, for example display it
              # print(img.format, file_path)
              if img.format in ("PNG","JPEG","JPG"):
                shutil.copyfile(file_path, os.path.join(dest_folder_path, file_name))
              # img.show()
              # break
      except IOError:
          print(f"Error opening file: {file_path}")
```
